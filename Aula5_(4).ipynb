#### Aplica o algoritmo para dividir o conjunto de dados em 2 subconjuntos: um para treinamento e outro para teste.

```
import matplotlib.pyplot as plt                           # Importa a biblioteca matplotlib.

from sklearn import datasets, svm, metrics                # Importa novamente as 3 funções da biblioteca SKlearn (maching learning).
from sklearn.model_selection import train_test_split      # Importa novamente a função que divide 1 conjunto de dados em 2 (treinamento e teste).

digits = datasets.load_digits()     # Carrega novamente o conjunto de imagens de dígitos (base prota da biblioteca SKlearning).

# Exibe o conjunto de dados de "target" (objetivo):

_, axes = plt.subplots(2, 4)
images_and_labels = list(zip(digits.images, digits.target))
for ax, (image, label) in zip(axes[0, :], images_and_labels[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Training: %i' % label)



n_samples = len(digits.images)                   # Obtem o tamanho do conjunto de dados.
data = digits.images.reshape((n_samples, -1))    # Reformata o conjunto de dados original como matriz.


# Reaplica o algoritmo de ML anterior:

classifier = svm.SVC(gamma=0.001)

X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.5, shuffle=False)

classifier.fit(X_train, y_train)

predicted = classifier.predict(X_test)

images_and_predictions = list(zip(digits.images[n_samples // 2:], predicted))
for ax, (image, prediction) in zip(axes[1, :], images_and_predictions[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Prediction: %i' % prediction)

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(y_test, predicted)))
disp = metrics.plot_confusion_matrix(classifier, X_test, y_test)
disp.figure_.suptitle("Confusion Matrix")
print("Confusion matrix:\n%s" % disp.confusion_matrix)

plt.show()
```

#### Exercício 2

```
# Importa funções das bibliotecas MatPlotLib, NumPy e SKLearning (matemáticas, plotagem e maching learning).

import matplotlib.pyplot as plt                             
from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```
# Importa diversos algoritmos de ML (classificadoras) da SKLearning.

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

```

```
# Cria um vetor (array) contendo os nomes das funções (algoritmos) da SKLearning. 
names = ["Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

# Cria um vetor (array) contendo os métodos classificadores (algoritmos) da SKLearning. 
classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]

# Aplica um calculo matemático (qualquer) para gerar um conjunto aleatório de dados (dataset).
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

# gera um vetor com 3 datasets a partir do calculo efetuado.
new_datasets = [make_moons(noise=0.3, random_state=0),
            make_circles(noise=0.2, factor=0.5, random_state=1),
            linearly_separable
            ]

# Cria uma área de exibição de uma figura.
figure = plt.figure(figsize=(54, 18))
i = 1
# iterate over datasets

# Para cada dataset criado, separa o conjunto de dados em treinamento e teste,
# e aplica os métodos classificadores da ML salvos no vetor.

for ds_cnt, ds in enumerate(new_datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.4, random_state=42)

    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    # xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    xx, yy = np.meshgrid(np.arange(x_min, x_max), np.arange(y_min, y_max))

    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    ax = plt.subplot(len(new_datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
               edgecolors='k')
    # Plot the testing points
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6,
               edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1

    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(new_datasets), len(classifiers) + 1, i)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)

        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, x_max]x[y_min, y_max].
        if hasattr(clf, "decision_function"):
            Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
        else:
            Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

        # Plot the training points
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
                   edgecolors='k')
        # Plot the testing points
        ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
                   edgecolors='k', alpha=0.6)

        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        if ds_cnt == 0:
            ax.set_title(name)
        ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
                size=15, horizontalalignment='right')
        i += 1
```

```
plt.tight_layout()
plt.show()
```

```
# Importa a função de plotagem da biblioteca matplotlib, a biblioteca NumPy, o 
# método de regressação linear (algoritmo de ML) da SKLeaning e os métpodos para
# aplicação de métricas para análise de resultados.

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)    # Carrega um dataset da SKLearning referente à dados sobre diabetes.
```

```
print(diabetes_X.shape)    # Exibe a dimensão da matriz de diabetes no eixo X (dados de entrada na função).
print(diabetes_y.shape)    # Exibe a dimensão da matriz de diabetes no eixo Y (dados de saída na função).
print(diabetes_X)          # Exibe uma amostra dos dados no eixo X (dados de entrada = treino e teste).
print(diabetes_y)          # Exibe uma amostra dos dados no eixo Y (dados de saída = target (glicose)).
```

```
diabetes_X = diabetes_X[:, np.newaxis, 2]     # Separa apenas uma coluna no eixo "X" para fazer a correlação com "y" (regressão linear).
print(diabetes_X[:10])                        # Exibe os 10 últimos elementos da coluna selecionada.
```

```
diabetes_X_train = diabetes_X[:-20]    # Reserva os últimos 20 elementos no eixo "X" para treino.
diabetes_X_test = diabetes_X[-20:]     # Reserva os antecedentes dos últimos 20 elementos no eixo "X" para teste.
diabetes_y_train = diabetes_y[:-20]    # Reserva os últimos 20 elementos no eixo "y" para treino.
diabetes_y_test = diabetes_y[-20:]     # Reserva os antecedentes dos últimos 20 elementos no eixo "y" para teste.
```

```
regr = linear_model.LinearRegression()            # Cria o objeto para aplicar o algoritmo de regressão linear (ML).
regr.fit(diabetes_X_train, diabetes_y_train)      # Aplica a função de ajuste nos dados de treino (algoritmo).
diabetes_y_pred = regr.predict(diabetes_X_test)   # Aplica a função de predição nos dados de test.
```

```
print('Coeficiente de determinação: %.2f'
      % r2_score(diabetes_y_test, diabetes_y_pred))  # Calcula o coeficiente de precisão entre a predição e o teste.
```

```
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')          # Aplica a função de distribuição dos dados de teste no gráfico.
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue', linewidth=3)  # Aplica a função de regressão linear na distribuição.

# Exibe o gráfico. 
plt.xticks(())
plt.yticks(())
plt.show()
```

```
# Calcula a equação da reta : a.x + b
a = regr.coef_           # Obtem a inclinação da reta no gráfico
b = regr.intercept_      # Obtem a intersecção da reta com o eixo "x".
print (a, '* x  + ', b)
```

### Agrupamento

#### K-Means

```
# Importa funções de bibliotecas.

import matplotlib.pyplot as plt         # Importa função de plotagem da matplotlib.
from sklearn.datasets import make_blobs # Importa algoritmo da SKLearning.

# Gera (inventa) um conjunto de dados aletórios agrupados.

# create dataset
X, y = make_blobs(
   n_samples=150, n_features=2,
   centers=3, cluster_std=0.5,
   shuffle=True, random_state=0
)


# Exibe os dados agrupados no gráfico.

# plot
plt.scatter(
   X[:, 0], X[:, 1],
   c='white', marker='o',
   edgecolor='black', s=50
)
plt.show()
```

```
from sklearn.cluster import KMeans     # Importa o algoritmo KMeans da biblioteca SK Learning.

km = KMeans(n_clusters=3)              # Aplica o algoritmo definindo a quantidade de grupos (n_clusters = 3).

y_km = km.fit_predict(X)               # Efetua a predição com base no eixo "X" (classificação não supervisionada, isso é, sem teste).
```

```
print(km.cluster_centers_)             # Calcula e exibe as coordenadas dos centroides dos grupos.
```

```
# Efetua a plotagem dos centroides calculados e seus respectivos grupos.

# plot the 3 clusters (3 grupos)

plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

# plot the centroids
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroides'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```
# Efetua a plotagem dos centroides calculados e seus respectivos grupos.

km = KMeans(n_clusters=4)     # Aplica o algoritmo definindo a quantidade de grupos (n_clusters = 3).

y_km = km.fit_predict(X)      # Efetua a predição com base no eixo "X" (classificação não supervisionada, isso é, sem teste).

print(km.cluster_centers_)    # Calcula e exibe as coordenadas dos centroides dos grupos.

# plot the 4 clusters (4 grupos)

plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

plt.scatter(
    X[y_km == 3, 0], X[y_km == 3, 1],
    s=50, c='pink',
    marker='v', edgecolor='black',
    label='cluster 4'
)

plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroides'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```
