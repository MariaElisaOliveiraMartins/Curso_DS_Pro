### Scatter plots (gráficos de pontos)

````
# Criando vetores aleatórios 

x = range(100)
y1_samples = np.random.rand(100)*3   # Cria primeira amostra a partir de números aleatórios.
y2_samples = np.random.rand(100)*4   # Cria segunda amostra a partir de números aleatórios.
y3_samples = np.random.rand(100)*5   # Cria terceira amostra a partir de números aleatórios.

# Cria-se a figura

plt.figure(figsize=(8,6))            # Define o tamanho da figura.
plt.scatter(x,y1_samples, marker='x', 
            color='blue', alpha=0.7, label='amostra y1')     # Exibe primeira amostra conforme parametrização.
plt.scatter(x,y2_samples, marker='o', 
            color='green', alpha=0.7, label='amostra y2')    # Exibe segunda amostra conforme parametrização.
plt.scatter(x,y3_samples, marker='s', 
            color='red', alpha=0.7, label='amostra y3')      # Exibe terceira amostra conforme parametrização.

plt.title('Gráfico scatter')     # Atribui um título ao gráfico.
plt.ylabel('Variável Y')         # Atribui um rótulo ao eixo Y.
plt.xlabel('Variável X')         # Atribui um rótulo ao eixo X.
plt.legend(loc='upper right')    # Exibe uma legenda.

plt.show()
````

````
# Implementa uma função em python para gerar amostras segundo uma distribuição normal. 

import numpy as np
from matplotlib import pyplot as plt
import math

# Implementação da função de distribuição normal:
def pdf(x, mu=0, sigma=1):
    """
    Calcula a função de densidade de distribuição normal
        
    """
    term1 = 1.0 / ( math.sqrt(2*np.pi) * sigma )
    term2 = np.exp( -0.5 * ( (x-mu)/sigma )**2 )
    return term1 * term2


x = np.arange(0, 100, 0.05)

# Gera as amostras utilizando a função implementada:
pdf1 = pdf(x, mu=5, sigma=2.5**0.5)    
pdf2 = pdf(x, mu=10, sigma=6**0.5)

# Aplica as amostras no gráfico:
plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.title('Funções de densidade de probabilidade')
plt.ylabel('p(x)')
plt.xlabel('Variável aleatória x')
plt.legend(['pdf1 ~ N(5,2.5)', 'pdf2 ~ N(10,6)'], loc='upper right')
plt.ylim([0,0.5])
plt.xlim([0,20])

plt.show()    # Exibe o gráfico.
````

````
# Apresentação de gráficos de barras (Histograma) definindo o tamanho das barras.

import numpy as np
import random
from matplotlib import pyplot as plt

data = np.random.normal(0, 20, 1000)  # Gera amostra randômica de distribuição normal. 

# fixed bin size
bins = np.arange(-100, 100, 5) # fixando o tamanho do bin (bin = caixa, conteiner, silo, armazem, pote, etc. => barra)

plt.xlim([min(data)-5, max(data)+5])  # Define os limites (dimensão) do histograma no eixo X.

plt.hist(data, bins=bins, alpha=0.5)  # Aplica a amostra (alpha = intensidade da coloração). 
plt.title('Histogramas')
plt.xlabel('X (tamanho do bin = 5)')  
plt.ylabel('Contagem')

plt.show()   # Exibe o gráfico.
````

````
# Apresentação de gráficos de barras (Histograma) definindo a quantidade de barras.

import numpy as np
import random
import math
from matplotlib import pyplot as plt

# Utiliza a função "linspace" para definir a quantidade de caixas (barras).

data = np.random.normal(0, 20, 1000) 
 
bins = np.linspace(math.ceil(min(data)), 
                   math.floor(max(data)),
                   20) # número fixo de bins (barras)

plt.xlim([min(data)-5, max(data)+5])

plt.hist(data, bins=bins, alpha=0.5)
plt.title('Histogramas')
plt.xlabel('X (20 bins)')
plt.ylabel('Contagem')

plt.show()
````

````
# Apresentação de histograma identifcando as amostras pela coloração.

import matplotlib.pyplot as plt

# Input data
green_data = [1, 2, 3]
blue_data = [3, 2, 1]
red_data = [2, 3, 3]
labels = ['barra 1', 'barra 2', 'barra 3']


pos = list(range(len(green_data))) 
width = 0.2 
    
fig, ax = plt.subplots(figsize=(8,6))   # Define o tamanho da figura.

# Define a coloração e o posicionamento das barras para representar as amostras:

plt.bar(pos, green_data, width,
                 alpha=0.5,
                 color='g',
                 label=labels[0])

plt.bar([p + width for p in pos], blue_data, width,
                 alpha=0.5,
                 color='b',
                 label=labels[1])
    
plt.bar([p + width*2 for p in pos], red_data, width,
                 alpha=0.5,
                 color='r',
                 label=labels[2])


ax.set_ylabel('y')
ax.set_title('Barras agrupadas')
ax.set_xticks([p + 1.5 * width for p in pos])
ax.set_xticklabels(labels)


plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(green_data + blue_data + red_data) * 1.5])


plt.legend(['verde', 'azul', 'vermelho'], loc='upper left')
plt.grid()
plt.show()
````

````
# Apresentação de histograma no formato "stack bar" (barras empilhadas).

import matplotlib.pyplot as plt

blue_data = [100,120,140]
red_data = [150,120,190]
green_data = [80,70,90]

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

bar_width = 0.5

bar_l = [i+1 for i in range(len(blue_data))] 

tick_pos = [i+(bar_width/2) for i in bar_l] 

###################
## Valores absolutos
###################

ax1.bar(bar_l, blue_data, width=bar_width,
        label='azul', alpha=0.5, color='b')
ax1.bar(bar_l, red_data, width=bar_width,
        bottom=blue_data, label='vermelho', alpha=0.5, color='r')
ax1.bar(bar_l, green_data, width=bar_width,
        bottom=[i+j for i,j in zip(blue_data,red_data)], label='verde', alpha=0.5, color='g')

plt.sca(ax1)
plt.xticks(tick_pos, ['categoria 1', 'categoria 2', 'categoria 3'])

ax1.set_ylabel("Contagem")
ax1.set_xlabel("")
plt.legend(loc='upper left')
plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
plt.grid()

# Giro os eixos X
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')

############
## Percetuais
############

totals = [i+j+k for i,j,k in zip(blue_data, red_data, green_data)]
blue_rel = [i / j * 100 for  i,j in zip(blue_data, totals)]
red_rel = [i / j * 100 for  i,j in zip(red_data, totals)]
green_rel = [i / j * 100 for  i,j in zip(green_data, totals)]

ax2.bar(bar_l, blue_rel, 
        label='azul', alpha=0.5, color='b', width=bar_width
        )
ax2.bar(bar_l, red_rel, 
        bottom=blue_rel, label='vermelho', alpha=0.5, color='r', width=bar_width
        )
ax2.bar(bar_l, green_rel, 
        bottom=[i+j for i,j in zip(blue_rel, red_rel)], 
        label='verde', alpha=0.5, color='g', width=bar_width
        )

plt.sca(ax2)
plt.xticks(tick_pos, ['categoria 1', 'categoria 2', 'categoria 3'])
ax2.set_ylabel("Percentagem")
ax2.set_xlabel("")

plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
plt.grid()

# rotate axis labels
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')

plt.show()
````

````
# Apresentação de histograma horizontal incluindo anotações/informações nas barras.

from matplotlib import pyplot as plt
import numpy as np

data = range(200, 225, 5)

bar_labels = ['a', 'b', 'c', 'd', 'e']

fig = plt.figure(figsize=(10,8))

# barras
y_pos = np.arange(len(data))
plt.yticks(y_pos, bar_labels, fontsize=16)
bars = plt.barh(y_pos, data,
         align='center', alpha=0.4, color='g')

# anotações e labels

for b,d in zip(bars, data):
    plt.text(b.get_width() + b.get_width()*0.08, b.get_y() + b.get_height()/2,
        '{0:.2%}'.format(d/min(data)), 
        ha='center', va='bottom', fontsize=12)

plt.xlabel('X', fontsize=14)
plt.ylabel('Y', fontsize=14)
t = plt.title('Gráfico de barras com texto', fontsize=18)
plt.ylim([-1,len(data)+0.5])
plt.vlines(min(data), -1, len(data)+0.5, linestyles='dashed')
plt.grid()

plt.show()
````

````
# Apresentação de amostras em BoxPlot e seus repectivos outliers.

import matplotlib.pyplot as plt
import numpy as np

all_data = [np.random.normal(0, std, 100) for std in range(1, 4)]  # Gera 3 amostras de distribuição aleatória normal.

fig = plt.figure(figsize=(8,6))    # Define o tamanho da figura.

# Aplica o BoxPlot às amostras:
bplot = plt.boxplot(all_data, 
            notch=False,  # notch shape 
            vert=True,   # vertical box aligmnent
            patch_artist=True)   # fill with color

# Casa a coloração com as amostras para apresentação no gráfico:
colors = ['pink', 'lightblue', 'lightgreen']  
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

# Ajusta a exibição definindo os rótulos e o título no eixo X:
plt.xticks([y+1 for y in range(len(all_data))], ['x1', 'x2', 'x3'])
plt.xlabel('x')
t = plt.title('Box plot com diferentes cores')

plt.show()    # Exibe o BoxPlot.
````

````
# Apresentação de amostras em BoxPlot no formato "violino" e no formato tradicional para efeito de comparação.

import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12,5))

all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]


# Ajusta amostra para o formato "violino":
axes[0].violinplot(all_data,
               showmeans=False,
               showmedians=True 
               )
axes[0].set_title('Violino')

# Ajusta amostra para o formato padrão:
axes[1].boxplot(all_data,
               )
axes[1].set_title('Box')

# linhas de grid
for ax in axes:
    ax.yaxis.grid(True) 
    ax.set_xticks([y+1 for y in range(len(all_data))], )
    ax.set_xlabel('xlabel')
    ax.set_ylabel('ylabel')

# Define rótulos no eixo X.
plt.setp(axes, xticks=[y+1 for y in range(len(all_data))], 
        xticklabels=['x1', 'x2', 'x3', 'x4'],
        )    
    
plt.show()   # Exibe os gráficos.
````

````
# Exibe um gráfico em 3 dimensões (3D) usando a matplotlib:

from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import pyplot as plt

# dados
mu_vec1 = np.array([0,0,0]) # mean vector
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]]) # covariance matrix

class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20)
class2_sample = np.random.multivariate_normal(mu_vec1 + 1, cov_mat1, 20)
class3_sample = np.random.multivariate_normal(mu_vec1 + 2, cov_mat1, 20)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')    # Cria a figura em 3D através do parâmetro "projection".
   
ax.scatter(class1_sample[:,0], class1_sample[:,1], class1_sample[:,2], 
           marker='x', color='blue', s=40, label='classe 1')
ax.scatter(class2_sample[:,0], class2_sample[:,1], class2_sample[:,2], 
           marker='o', color='green', s=40, label='classe 2')
ax.scatter(class3_sample[:,0], class3_sample[:,1], class3_sample[:,2], 
           marker='^', color='red', s=40, label='classe 3')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('Gráfico de pontos 3D')
     
plt.show()
````

##### Criando um cubo contendo alguns dados

````
# Código que implementa um cubo em torno dos pontos a serem evidenciados no gráfico 3D (copiado da Internet):

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from itertools import product, combinations
fig = plt.figure(figsize=(7,7))
ax = fig.gca(projection='3d')

# Plot Points

# amostras dentro do cubo
X_inside = np.array([[0,0,0],[0.2,0.2,0.2],[0.1, -0.1, -0.3]])

X_outside = np.array([[-1.2,0.3,-0.3],[0.8,-0.82,-0.9],[1, 0.6, -0.7],
                      [0.8,0.7,0.2],[0.7,-0.8,-0.45],[-0.3, 0.6, 0.9],
                      [0.7,-0.6,-0.8]])

for row in X_inside:
    ax.scatter(row[0], row[1], row[2], color="r", s=50, marker='^')

for row in X_outside:    
    ax.scatter(row[0], row[1], row[2], color="k", s=50)

# Plot Cube
h = [-0.5, 0.5]
for s, e in combinations(np.array(list(product(h,h,h))), 2):
    if np.sum(np.abs(s-e)) == h[1]-h[0]:
        ax.plot3D(*zip(s,e), color="g")
        
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_zlim(-1.5, 1.5)

plt.show()
````

##### Plotando uma superfície colorida

````
# Código que implementa superfície em 3D (copiado da Internet):

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure(figsize=(15,9))
ax = fig.gca(projection='3d')

# Data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# plota a superfície  (obs: cmap=cm.coolwarm => exibição multicolorido)
surf = ax.plot_surface(X, Y, Z, cmap=cm.Blues,
                       linewidth=0, antialiased=False)

# Customizar eixo z
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Barra de cores
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
````
